```python
#Getting the categories of the videos using json file 
#[https://www.kaggle.com/venky73/youtube-trending-videos-analysis/data?select=US_category_id.json]
import re
import json
k = open(r"C:\Users\rups_\OneDrive - Oklahoma A and M System\MS BAnDS\Spring 2022\MSIS 5223-Programming\homework\msis-5223-deliverable-2-white-walkers\data\US_category_id.json").read()
Category_list = re.findall(r"title\": \"(.*)\"", k)
categories = {i+1:Category_list[i] for i in range(len(Category_list))}
categories
```

Gathering final views,likes,comments for each trending videos

```python
#df_finalViews stores the last occurance of VideoID
df_finalViews = df.sort_values('view_count', ascending=False).drop_duplicates(['video_id'])
df_finalViews.sample(5)
```

```python
#df_firstViews stores the data of a video appearing into trending videos for the first time
df_firstViews = df.sort_values('view_count', ascending=True).drop_duplicates(['video_id'])
df_firstViews.sample(5)
```

```python
#Fetching category to the three subset dataframes
def fetch_video_category(value):
    return categories[value]
df_finalViews["Video_Category"]=df_finalViews.categoryId.apply(fetch_video_category)
df_firstViews["Video_Category"]=df_firstViews.categoryId.apply(fetch_video_category)
df["Video_Category"]=df.categoryId.apply(fetch_video_category)

df.sample(5)
```

```python
#Removing unnecessary columns
df_finalViews.drop(["channelId","thumbnail_link","ratings_disabled","comments_disabled","description","tags"], inplace=True, axis=1)
df_firstViews.drop(["channelId","thumbnail_link","ratings_disabled","comments_disabled","description","tags"], inplace=True, axis=1)
df.drop(["channelId","thumbnail_link","ratings_disabled","comments_disabled","description","tags"], inplace=True, axis=1)
```

```python
df.sample(5)
```

```python
#View Analysis for each Category
df_Views_Analysis = df_finalViews.groupby('Video_Category')["view_count"].sum().reset_index()
df_Like = df_finalViews.groupby('Video_Category')["likes"].sum().reset_index()
df_Dislike = df_finalViews.groupby('Video_Category')["dislikes"].sum().reset_index()
df_Comments = df_finalViews.groupby('Video_Category')["comment_count"].sum().reset_index()
df_Views_Analysis["Likes"] = df_Like.likes
df_Views_Analysis["Dislikes"] = df_Dislike.dislikes
df_Views_Analysis
```

```python
#Response percentage=100*(like+dislikes)/total view
#Likes percentage=100*(likes)/total view
#Dislikes percentage=100*(dislikes)/total view
#Comments percentage=100*(comments)/total view
df_Views_Analysis["Response_Percentage"] = ( df_Views_Analysis.Likes + df_Views_Analysis.Dislikes ) * 100 /df_Views_Analysis.view_count
df_Views_Analysis["LikesPercentage"] = df_Views_Analysis.Likes*100/df_Views_Analysis.view_count
df_Views_Analysis["DislikesPercentage"] = df_Views_Analysis.Dislikes*100/df_Views_Analysis.view_count
df_Views_Analysis["Comments"] = df_Comments["comment_count"]
df_Views_Analysis["CommentsPercentage"] = df_Views_Analysis.Comments*100/df_Views_Analysis.view_count
df_Views_Analysis
```

View count does only give a surface level understanding of how popular a video could be, because people might have just viwed the video to give it a dislike. Visualizing the count of likes, and dislikes among different categories might give more understanding of which categories of videos people like and dislike the most.

```python
#Categories vs Likes
# Set the width and height of the figure
plt.figure(figsize=(20,10))

# Add title
plt.title("Likes Count vs Type of Videos")
plt.xticks(rotation=90)
sns.barplot(y=df_Views_Analysis.Video_Category, x=df_Views_Analysis.Likes, orient = 'h')

# Add label for vertical axis
plt.ylabel("Likes Count")
```

```python
#Categories vs DisLikes
# Set the width and height of the figure
plt.figure(figsize=(20,10))

# Add title
plt.title("Dislikes Count vs Type of Videos")
plt.xticks(rotation=90)
sns.barplot(y=df_Views_Analysis.Video_Category, x=df_Views_Analysis.Dislikes, orient = 'h')

# Add label for vertical axis
plt.ylabel("Dislikes Count")
```

Similar to the observation for views, it looks like People & Blogs, Drama, and Action/Adventure- these three categories outnumber other categories in count of likes and dislikes. This makes sense as a video can go viral in both positive and negative sense. Therefore we created another metric called response percentage that includes both the number of likes and dislikes and normalized it with the number of views. This will dictate how engaging the videos are to the viewers. 


```python
df_Views_Analysis["Response_Percentage"] = ( df_Views_Analysis.Likes + df_Views_Analysis.Dislikes ) * 100 /df_Views_Analysis.view_count
df_Views_Analysis
```

```python
#Categories vs Response %
# Set the width and height of the figure
plt.figure(figsize=(20,10))

# Add title
plt.title("response % vs Type of Videos")
plt.xticks(rotation=90)
sns.barplot(y=df_Views_Analysis.Video_Category, x=df_Views_Analysis.Response_Percentage)

# Add label for vertical axis
plt.ylabel("Response percentages")
```

Doumentary, Horror, and People & blogs are the three categories with highest response percentage. Let's randomly see what those videos are

No. of days took to be in trending

```python
df_firstViews["JoinedTrending"] = pd.to_datetime(df_firstViews.trending_date.str.split(',\s*').str[0])
df_firstViews["UploadedTime"] = pd.to_datetime(df_firstViews.publishedAt.str.split(',\s*').str[0])
df_firstViews["DaysTakenToBeOnTrending"] = (df_firstViews["JoinedTrending"] - df_firstViews["UploadedTime"]).dt.days
df_firstViews['DaysTakenToBeOnTrending'] = df_firstViews['DaysTakenToBeOnTrending'].apply(lambda x: 0 if x == -1 else x)
df_firstViews.sample(10)
```

```python
#how many days it takes to be viral

# Set the width and height of the figure
plt.figure(figsize=(20,10))

# Add title
plt.title("Average no.of days took to be in trending")
plt.xlabel
plt.xticks(rotation=90)

sns.barplot(x=df_firstViews.Video_Category, y=df_firstViews.DaysTakenToBeOnTrending)

# Add label for vertical axis
plt.ylabel("Avg no.of days")
```
